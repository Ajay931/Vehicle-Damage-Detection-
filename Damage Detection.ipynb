```python
#Labelling all images
def label(group):
    if group == 'bumper_dent':
        return np.array([0,0,0,0,0,0,0,1])
    if group == 'bumper_scratch':
        return np.array([0,0,0,0,0,0,1,0])
    if group == 'door_dent':
        return np.array([0,0,0,0,0,1,0,0])
    if group == 'door_scratch':
        return np.array([0,0,0,0,1,0,0,0])
    if group == 'glass_shatter':
        return np.array([0,0,0,1,0,0,0,0])
    if group == 'head_lamp':
        return np.array([0,0,1,0,0,0,0,0])
    if group == 'tail_lamp':
        return np.array([0,1,0,0,0,0,0,0])
    if group == 'unknown':
        return np.array([1,0,0,0,0,0,0,0])
```

```python
train_img = []
train_lab = []
test_img = []
test_lab = []
#All images
images = []
```

```python
import os
from PIL import Image
DIR = './'
```

```python
for i in range(1594):
    path = os.path.join(DIR,dataset.iloc[i]['image'])
    img = Image.open(path)
    cls = dataset.iloc[i]['class']  
    lab = label(cls)
    images.append([np.array(img),lab])
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.layers. normalization import BatchNormalization
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size = (3,3), activation = 'relu', input_shape = (224,224,3)))
model.add(MaxPooling2D(pool_size = (2,2)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(96, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(96, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(128, activation='relu'))
model.add(Dense(8, activation = 'softmax'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
```

```python
model.fit(train_img, train_lab, batch_size = 32, epochs = 10, verbose = 1)
```

```python
loss, acc = model.evaluate(test_img, test_lab, verbose = 0)
print(acc*100)
```
